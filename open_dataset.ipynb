```python
# scripts/extract_catalonia_csv.py

import xarray as xr
import pandas as pd

# Load the NetCDF datacube
ds = xr.open_dataset("IberFire.nc")

# Optional: inspect available variables and coordinates
print(ds)

# Define bounding box for Catalonia (approx.)
lat_bounds = (40.5, 43.0)
lon_bounds = (0.0, 3.5)

# Subset dataset
ds_cat = ds.sel(lat=slice(*lat_bounds), lon=slice(*lon_bounds))

# Convert to DataFrame (flattened)
df = ds_cat.to_dataframe().reset_index()

# Optional: filter for years 2021–2023
df["date"] = pd.to_datetime(df["date"])
df = df[(df["date"] >= "2021-01-01") & (df["date"] <= "2023-12-31")]

# Save to CSV
df.to_csv("data/iberfire_catalonia_2021_2023.csv", index=False)
print("✅ Saved filtered Catalonia dataset.")
```
